### ğŸ”„ Flashback #1: First Law of Thermodynamics (Energy Conservation)

**Analogy Time - Pocket Money Scene:**

Soch bhai, tere paas pocket money aati hai (Energy In), aur tu kuch kharche karta hai (Energy Out). Jo bacha wo teri savings (Energy Stored).

> **Jo aaya - Jo gaya = Jo bacha**

Physics me bhi same scene hai:

$$Q_{in} - W_{out} = \Delta E_{system}$$

**For a CYCLE (jab system wapas same state me aa jaaye):**

Cycle complete hone ke baad system wapas wohi hai jaha se shuru hua, matlab **Î”E = 0**

$$Q_{net} = W_{net}$$

Ya phir Heat Engine ke liye:

$$Q_{in} - Q_{out} = W_{net}$$

### âš¡ Flashback #2: Efficiency (Î·) - Kitna Kaam ka Hai?

**Analogy Time - Exam Marks Scene:**

Bhai, tune 10 ghante padhai ki (Input), lekin exam me sirf 6 ghante wali cheezein kaam aayi (Useful Output). Teri efficiency = 60%!

> **Efficiency = Kya mila / Kya diya Ã— 100%**

**Heat Engine ke liye:**

$$\eta = \frac{W_{net}}{Q_{in}} = \frac{\text{Kaam nikla}}{\text{Heat daali}}$$

**Furnace/Boiler ke liye:**

$$\eta = \frac{Q_{out}}{Q_{in}} = \frac{\text{Ghar ko mili heat}}{\text{Fuel me thi heat}}$$

### ğŸŒ¡ï¸ Flashback #3: COP (Coefficient of Performance) - "Paisa Wasool" Factor

**Analogy Time - Zomato Delivery Scene:**

Tu 50 Rs delivery charge deta hai, but 500 Rs ka khana aata hai. Tere paise ka 10x value mila!

Heat Pump/AC me bhi same scene hai. Tu electricity (Work) deta hai, but usse zyada heat transfer hoti hai!

**Heat Pump ke liye (Heating mode):**

$$COP_{HP} = \frac{Q_H}{W_{in}} = \frac{\text{Ghar ko mili heat}}{\text{Bijli ka kaam}}$$

**AC/Refrigerator ke liye (Cooling mode):**

$$COP_{R} = \frac{Q_L}{W_{in}} = \frac{\text{Room se nikali heat}}{\text{Bijli ka kaam}}$$

**Important:** COP > 1 ho sakta hai! (Unlike efficiency jo max 100%)

### ğŸ§® Step-by-Step Solution:

---

**STEP 1: First Law Apply Karo (Cycle ke liye)**

Cycle ke liye: $Q_{in} - Q_{out} = W_{net}$

---

**STEP 2: Efficiency Formula se Q_in Nikalo**

$$\eta = \frac{W_{net}}{Q_{in}}$$

Rearrange karo:

$$Q_{in} = \frac{W_{net}}{\eta}$$

Values daalo:

$$Q_{in} = \frac{50,000 \text{ kW}}{0.21}$$

</div>

<div style="background-color: #161b22; border-left: 5px solid #a371f7; padding: 15px; margin: 15px 0; color: #c9d1d9;">

**ğŸ’¡ Important Note:**

kW matlab kJ/s hai! Power = Energy per unit time

So, $Q_{in}$ bhi kJ/s me aayega (rate of heat input)

</div>

<div style="background-color: #0d1117; padding: 20px; color: #c9d1d9;">

$$\dot{Q}_{in} = \frac{50,000}{0.21} = 238,095.24 \text{ kJ/s}$$

---

**STEP 3: Heat Input se Mass Flow Rate Nikalo**

Ab scene yeh hai ki heat aa raha hai fuel jalane se:

$$\dot{Q}_{in} = \dot{m} \times H.V.$$

Jaha:
- $\dot{m}$ = mass flow rate (kg/s) - yahi nikalna hai!
- H.V. = Heating Value = 42,000 kJ/kg

Rearrange:

$$\dot{m} = \frac{\dot{Q}_{in}}{H.V.}$$

$$\dot{m} = \frac{238,095.24}{42,000}$$

$$\dot{m} = 5.669 \text{ kg/s}$$

</div>

<div style="background-color: #0d1117; padding: 20px; color: #c9d1d9;">

**STEP 4: Mass Flow Rate se Volume Flow Rate Nikalo**

Humne kg/s nikala, but question L/s me maang raha hai.

Yaad karo: $\rho = \frac{m}{V}$ matlab $V = \frac{m}{\rho}$

Rates ke liye:

$$\dot{V} = \frac{\dot{m}}{\rho}$$

Given: $\rho = 0.8 \text{ kg/L}$

$$\dot{V} = \frac{5.669 \text{ kg/s}}{0.8 \text{ kg/L}}$$

$$\boxed{\dot{V} = 7.09 \text{ L/s}}$$

</div>

<div style="background-color: #161b22; border-left: 5px solid #3fb950; padding: 20px; margin: 20px 0; color: #c9d1d9; border-radius: 0 10px 10px 0;">

### ğŸ§® Solution - Part A: Gas Furnace

---

**Heating Requirement:** 1 therm = 1,05,500 kJ (ghar ko itni heat chahiye)

**Furnace Efficiency:** 97% = 0.97

Furnace me scene yeh hai:

$$\eta_{furnace} = \frac{Q_{out}}{Q_{in}}$$

Matlab:

$$Q_{in} = \frac{Q_{out}}{\eta} = \frac{1,05,500}{0.97}$$

$$Q_{in} = 1,08,763 \text{ kJ}$$

---

**Cost Calculation:**

Ab yeh â‚¹60/therm hai, to pehle therms me convert karo:

$$\text{Therms needed} = \frac{1,08,763}{1,05,500} = 1.031 \text{ therms}$$

$$\text{Cost} = 1.031 \times 60 = \boxed{â‚¹61.85}$$

### ğŸ§® Solution - Part B: Heat Pump

---

**Same Heating Requirement:** $Q_H$ = 1,05,500 kJ (ghar ko itni heat chahiye)

**Heat Pump COP:** 3.5

Heat Pump me scene yeh hai:

$$COP_{HP} = \frac{Q_H}{W_{in}}$$

Rearrange:

$$W_{in} = \frac{Q_H}{COP} = \frac{1,05,500}{3.5}$$

$$W_{in} = 30,143 \text{ kJ}$$

**ğŸ’¡ Unit Conversion Alert!**

Electricity ka rate kWh me hai, but humara answer kJ me hai!

$$1 \text{ kWh} = 1 \text{ kW} \times 1 \text{ hour} = 1000 \text{ J/s} \times 3600 \text{ s} = 3600 \text{ kJ}$$
