### ğŸ”¥ BFS Template - Yaad Kar Le!

```python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    visited.add(start)
    
    while queue:
        node = queue.popleft()  # FIFO - pehle aaya pehle gaya
        print(node)  # Process karo
        
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
```

### ğŸ”¥ DFS Template - Yaad Kar Le!

```python
# Recursive DFS (Zyada clean)
def dfs(graph, node, visited):
    visited.add(node)
    print(node)  # Process karo
    
    for neighbor in graph[node]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

# Iterative DFS (Stack use karke)
def dfs_iterative(graph, start):
    visited = set()
    stack = [start]
    
    while stack:
        node = stack.pop()  # LIFO - last aaya pehle gaya
        if node not in visited:
            visited.add(node)
            print(node)
            for neighbor in graph[node]:
                stack.append(neighbor)
```

### ğŸ§  Kab BFS, Kab DFS?

| Situation | Use This | Reason |
|-----------|----------|--------|
| Shortest path (unweighted) | **BFS** | Level-wise jaata hai |
| Path exists ya nahi? | **DFS** | Simpler, recursive |
| Connected components | **DFS** | Easy to implement |
| Cycle detection | **DFS** | Backtracking easy |
| Level order traversal | **BFS** | Nature hi level-wise hai |
| Maze/Grid problems | **Both work** | DFS thoda easy likhna |

â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
</pre>

## ğŸ’» Solution 1: DFS Approach (Recursive)

```python
class Solution:
    def validPath(self, n: int, edges: List[List[int]], source: int, destination: int) -> bool:
        # Edge case: source hi destination hai
        if source == destination:
            return True
        
        # Step 1: Adjacency List banao
        from collections import defaultdict
        graph = defaultdict(list)
        
        for u, v in edges:
            graph[u].append(v)  # u se v ja sakte
            graph[v].append(u)  # v se u bhi (undirected)
        
        # Step 2: DFS karo
        visited = set()
        
        def dfs(node):
            # Base case: destination mil gaya!
            if node == destination:
                return True
            
            visited.add(node)
            
            # Sab neighbors check karo
            for neighbor in graph[node]:
                if neighbor not in visited:
                    if dfs(neighbor):  # Agar kisi path se mila
                        return True
            
            return False  # Kisi path se nahi mila
        
        return dfs(source)
```

## ğŸ’» Solution 2: BFS Approach

```python
class Solution:
    def validPath(self, n: int, edges: List[List[int]], source: int, destination: int) -> bool:
        if source == destination:
            return True
        
        # Adjacency List
        from collections import defaultdict, deque
        graph = defaultdict(list)
        
        for u, v in edges:
            graph[u].append(v)
            graph[v].append(u)
        
        # BFS
        visited = set([source])
        queue = deque([source])
        
        while queue:
            node = queue.popleft()
            
            if node == destination:  # Mil gaya!
                return True
            
            for neighbor in graph[node]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)
        
        return False  # Queue khatam, nahi mila
```

## ğŸ’» Solution 3: Union Find (Advanced)

> Bhai, yeh thoda advanced hai but **bahut powerful** technique hai. Isko samajh le!

```python
class Solution:
    def validPath(self, n: int, edges: List[List[int]], source: int, destination: int) -> bool:
        # Union Find / Disjoint Set
        parent = list(range(n))  # Har node apna khud ka parent
        
        def find(x):
            if parent[x] != x:
                parent[x] = find(parent[x])  # Path compression
            return parent[x]
        
        def union(x, y):
            px, py = find(x), find(y)
            if px != py:
                parent[px] = py  # Merge karo
        
        # Sab edges ko union karo
        for u, v in edges:
            union(u, v)
        
        # Check if same component
        return find(source) == find(destination)
```

**Step 3:** "Toh main har person ka in-degree aur out-degree count karunga"

**ğŸ’¡ Optimization:** "Wait! In-degree - Out-degree = n-1 wala hi judge hoga!"

Kyun?

- Kisi aur ka kabhi n-1 nahi ho sakta!

## ğŸ’» Solution 1: Two Arrays (Clear Approach)

```python
class Solution:
    def findJudge(self, n: int, trust: List[List[int]]) -> int:
        if n == 1:
            return 1  # Akela insaan khud judge hai
        
        in_degree = [0] * (n + 1)   # Kitne log is pe trust karte
        out_degree = [0] * (n + 1)  # Yeh kitne pe trust karta
        
        for a, b in trust:
            out_degree[a] += 1  # a trusts someone
            in_degree[b] += 1   # b is trusted by someone
        
        for person in range(1, n + 1):
            if in_degree[person] == n - 1 and out_degree[person] == 0:
                return person
        
        return -1
```

## ğŸ’» Solution 2: Single Array (Space Optimized)

```python
class Solution:
    def findJudge(self, n: int, trust: List[List[int]]) -> int:
        if n == 1:
            return 1
        
        # trust_score = in_degree - out_degree
        trust_score = [0] * (n + 1)
        
        for a, b in trust:
            trust_score[a] -= 1  # Kisi pe trust kiya = -1
            trust_score[b] += 1  # Kisi ne trust kiya = +1
        
        # Judge ka score = (n-1) - 0 = n-1
        for person in range(1, n + 1):
            if trust_score[person] == n - 1:
                return person
        
        return -1
```

### ğŸ”§ Grid Traversal Template (YAAD KAR!)

```python
# 4 directions: up, down, left, right
directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]

# Ya ek aur style:
for dr, dc in [(0,1), (0,-1), (1,0), (-1,0)]:
    new_row = row + dr
    new_col = col + dc
    
    # Boundary check
    if 0 <= new_row < rows and 0 <= new_col < cols:
        # Valid cell hai, kuch karo
        pass
```

## ğŸ’» Optimal Solution: DFS

```python
class Solution:
    def floodFill(self, image: List[List[int]], sr: int, sc: int, color: int) -> List[List[int]]:
        rows, cols = len(image), len(image[0])
        original_color = image[sr][sc]
        
        # Edge case: Same color = kuch karna nahi
        if original_color == color:
            return image
        
        def dfs(r, c):
            # Boundary check + color check
            if r < 0 or r >= rows or c < 0 or c >= cols:
                return
            if image[r][c] != original_color:
                return
            
            # Fill with new color
            image[r][c] = color
            
            # Visit all 4 directions
            dfs(r + 1, c)  # down
            dfs(r - 1, c)  # up
            dfs(r, c + 1)  # right
            dfs(r, c - 1)  # left
        
        dfs(sr, sc)
        return image
```

## ğŸ’» BFS Solution (Alternative)

```python
from collections import deque

class Solution:
    def floodFill(self, image: List[List[int]], sr: int, sc: int, color: int) -> List[List[int]]:
        rows, cols = len(image), len(image[0])
        original_color = image[sr][sc]
        
        if original_color == color:
            return image
        
        queue = deque([(sr, sc)])
        image[sr][sc] = color
        
        while queue:
            r, c = queue.popleft()
            
            for dr, dc in [(0,1), (0,-1), (1,0), (-1,0)]:
                nr, nc = r + dr, c + dc
                
                if 0 <= nr < rows and 0 <= nc < cols and image[nr][nc] == original_color:
                    image[nr][nc] = color
                    queue.append((nr, nc))
        
        return image
```

> Jab land cell water ya boundary se touch kare

**Step 2:** "Har land cell ke 4 sides hain"
> Har side check karo - agar neighbor water/boundary hai toh +1

**Step 3:** "Approach Options:"
> - **Option A:** Count edges touching water (Direct)
> - **Option B:** Total sides - shared sides (Math trick)

**ğŸ’¡ Option B is smarter:**
> - Har land cell = 4 sides
> - But jab do land adjacent hain = 2 sides share (cancel)
> - Formula: **4 Ã— lands - 2 Ã— shared_edges**

## ğŸ’» Solution 1: Count Each Edge (Direct)

```python
class Solution:
    def islandPerimeter(self, grid: List[List[int]]) -> int:
        rows, cols = len(grid), len(grid[0])
        perimeter = 0
        
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == 1:  # Land cell mila
                    # Check all 4 sides
                    # Top: boundary ya water?
                    if r == 0 or grid[r-1][c] == 0:
                        perimeter += 1
                    # Bottom
                    if r == rows-1 or grid[r+1][c] == 0:
                        perimeter += 1
                    # Left
                    if c == 0 or grid[r][c-1] == 0:
                        perimeter += 1
                    # Right
                    if c == cols-1 or grid[r][c+1] == 0:
                        perimeter += 1
        
        return perimeter
```

## ğŸ’» Solution 2: Math Trick (Elegant)

```python
class Solution:
    def islandPerimeter(self, grid: List[List[int]]) -> int:
        rows, cols = len(grid), len(grid[0])
        land_count = 0
        shared_edges = 0
        
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == 1:
                    land_count += 1
                    
                    # Sirf right aur down check karo (avoid double counting)
                    if r + 1 < rows and grid[r+1][c] == 1:
                        shared_edges += 1
                    if c + 1 < cols and grid[r][c+1] == 1:
                        shared_edges += 1
        
        # Formula: Each land has 4 sides, but shared edges cancel
        return 4 * land_count - 2 * shared_edges
```

## ğŸ’» Optimal Solution: DFS with Area Count

```python
class Solution:
    def maxAreaOfIsland(self, grid: List[List[int]]) -> int:
        rows, cols = len(grid), len(grid[0])
        max_area = 0
        
        def dfs(r, c):
            # Boundary check + land check
            if r < 0 or r >= rows or c < 0 or c >= cols or grid[r][c] == 0:
                return 0
            
            # Mark as visited by changing to 0 (sink the island)
            grid[r][c] = 0
            
            # Current cell (1) + area from all 4 directions
            area = 1  # Current cell
            area += dfs(r + 1, c)  # down
            area += dfs(r - 1, c)  # up
            area += dfs(r, c + 1)  # right
            area += dfs(r, c - 1)  # left
            
            return area
        
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == 1:  # New island found!
                    island_area = dfs(r, c)
                    max_area = max(max_area, island_area)
        
        return max_area
```

```python
class Solution:
    def maxAreaOfIsland(self, grid: List[List[int]]) -> int:
        rows, cols = len(grid), len(grid[0])
        visited = set()
        max_area = 0
        
        def dfs(r, c):
            if (r < 0 or r >= rows or c < 0 or c >= cols or 
                grid[r][c] == 0 or (r, c) in visited):
                return 0
            
            visited.add((r, c))
            
            return (1 + dfs(r+1, c) + dfs(r-1, c) + 
                       dfs(r, c+1) + dfs(r, c-1))
        
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == 1 and (r, c) not in visited:
                    max_area = max(max_area, dfs(r, c))
        
        return max_area
```

> Room i â†’ Room j if room i has key j

**Step 2:** "Sab rooms visit ho sakte hain?"
> = Can we reach all nodes starting from node 0?

**Step 3:** "Yeh toh simple DFS/BFS traversal hai!"
> Start from 0, traverse, count visited = n means all reachable

**ğŸ’¡ Key Insight:** rooms array khud adjacency list hai!
> rooms[i] = list of rooms accessible from room i

## ğŸ’» Solution 1: DFS

```python
class Solution:
    def canVisitAllRooms(self, rooms: List[List[int]]) -> bool:
        visited = set()
        
        def dfs(room):
            visited.add(room)
            
            for key in rooms[room]:  # rooms[room] = keys in this room
                if key not in visited:
                    dfs(key)
        
        dfs(0)  # Start from room 0
        
        return len(visited) == len(rooms)  # Sab visit ho gaye?
```

## ğŸ’» Solution 2: BFS

```python
from collections import deque

class Solution:
    def canVisitAllRooms(self, rooms: List[List[int]]) -> bool:
        visited = set([0])  # Room 0 visited
        queue = deque([0])
        
        while queue:
            room = queue.popleft()
            
            for key in rooms[room]:
                if key not in visited:
                    visited.add(key)
                    queue.append(key)
        
        return len(visited) == len(rooms)
```

## ğŸ’» Solution 3: Iterative DFS (Stack)

```python
class Solution:
    def canVisitAllRooms(self, rooms: List[List[int]]) -> bool:
        visited = set()
        stack = [0]
        
        while stack:
            room = stack.pop()
            
            if room not in visited:
                visited.add(room)
                
                for key in rooms[room]:
                    if key not in visited:
                        stack.append(key)
        
        return len(visited) == len(rooms)
```
