<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #58a6ff;">
<strong style="color: #58a6ff;">üìö CONCEPT: Phase</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
A <strong>phase</strong> is a quantity of matter that is <strong>homogeneous throughout</strong> in:
</p>
<ol style="color: #c9d1d9;">
<li><strong>Chemical Composition</strong> - Same chemical makeup everywhere</li>
<li><strong>Physical Structure</strong> - All solid, OR all liquid, OR all vapor (gas)</li>
</ol>
</div>

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #3fb950;">üí° Real-Life Analogy</h3>
<p style="color: #c9d1d9;">
Think of phases like different <strong>states of WhatsApp groups</strong>:
</p>
<ul style="color: #c9d1d9;">
<li><strong>Solid Phase</strong> = Dead group (no movement, everyone stuck in place) üßä</li>
<li><strong>Liquid Phase</strong> = Active group (people moving, chatting, flowing) üíß</li>
<li><strong>Vapor Phase</strong> = Chaotic group (everyone everywhere, maximum freedom) üí®</li>
</ul>
<p style="color: #8b949e; font-style: italic;">A system can have multiple phases coexisting - like ice cubes (solid) floating in water (liquid)!</p>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #58a6ff;">
<strong style="color: #58a6ff;">üìö CONCEPT: Pure Substance</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
A <strong>pure substance</strong> is uniform and invariable in <strong>chemical composition</strong>.
</p>
<p style="color: #c9d1d9;">
‚úÖ Can exist in <strong>more than one phase</strong><br>
‚úÖ But chemical composition must be <strong>same in each phase</strong>
</p>
</div>

<div style="background: linear-gradient(to right, #161b22, #0d1117); padding: 15px; border-left: 5px solid #d29922; border-radius: 8px;">
<h2 style="color: #d29922; margin:0;">‚ö° 4.1 Fixing the State (The State Principle)</h2>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #58a6ff;">
<strong style="color: #58a6ff;">üìö CONCEPT: Intensive State</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
The <strong>intensive state</strong> of a closed system at equilibrium = Its condition described by values of its <strong>intensive thermodynamic properties</strong>.
</p>
</div>

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #d29922;">üî• Here's the Magic...</h3>
<p style="color: #c9d1d9;">
Not all properties are <strong>independent</strong>! Some properties are connected to others.
</p>
<p style="color: #c9d1d9;">
You don't need to know <strong>ALL</strong> properties to describe a state. Just know a few <strong>independent</strong> ones, and the rest are automatically determined!
</p>
<p style="color: #3fb950; font-weight: bold; margin-top: 10px;">
Question: How many independent properties do we need? ü§î
</p>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #d29922;">
<strong style="color: #d29922;">‚ö†Ô∏è THE STATE PRINCIPLE (a.k.a. State Postulate)</strong>
<p style="color: #c9d1d9; margin-top: 10px; font-size: 16px;">
The number of <strong>independent properties</strong> needed to fix the state = <strong>One property for each way a system's energy can be varied independently</strong>
</p>
</div>

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #58a6ff;">üßÆ Breaking Down the Formula</h3>
<p style="color: #c9d1d9;">
Energy of a system can change through:
</p>
<ol style="color: #c9d1d9; line-height: 1.8;">
<li><strong>Heat Transfer (Q)</strong> ‚Üí 1 independent property</li>
<li><strong>Work Interactions (W)</strong> ‚Üí 1 independent property for each type of work</li>
</ol>
<p style="color: #3fb950; font-weight: bold; margin-top: 15px;">
üìê Formula: <span style="background-color: #161b22; padding: 5px 10px; border-radius: 5px;">Number of Independent Properties = 1 + (Number of Relevant Work Modes)</span>
</p>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #3fb950;">
<strong style="color: #3fb950;">‚úÖ CONCEPT: Simple System</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
A <strong>simple system</strong> = System where energy can be altered by work in <strong>only ONE way</strong> during quasiequilibrium processes.
</p>
<p style="color: #c9d1d9; margin-top: 10px;">
For Simple Systems:<br>
‚úÖ 1 property for heat transfer<br>
‚úÖ 1 property for single work mode<br>
</p>
<p style="color: #3fb950; font-weight: bold; margin-top: 10px; font-size: 18px;">
üìå Total = <span style="background-color: #0d1117; padding: 5px 15px; border-radius: 5px;">2 INDEPENDENT PROPERTIES</span>
</p>
</div>

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #3fb950;">üéØ Types of Simple Systems</h3>

| Type | Work Mode | Work Formula | Example |
|------|-----------|--------------|----------|
| **Simple Compressible** | Volume Change | W = ‚à´p dV | Piston-cylinder |
| **Simple Elastic** | Stretching | W = ‚à´F dx | Rubber band |
| **Simple Magnetic** | Magnetic field change | W = ‚à´H dM | Electromagnet |

<p style="color: #8b949e; margin-top: 10px; font-style: italic;">In this course, we mostly deal with <strong style="color: #3fb950;">Simple Compressible Systems</strong>!</p>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px dashed #f85149;">
<h3 style="color: #f85149; margin-top:0;">üö´ Important Exclusion!</h3>
<p style="color: #c9d1d9;">
Properties like <strong>velocity</strong> and <strong>elevation</strong> are <strong>NOT</strong> considered here!
</p>
<p style="color: #c9d1d9;">
Why? Because:
</p>
<ul style="color: #c9d1d9;">
<li>They are <strong>mechanical properties</strong>, not thermodynamic properties</li>
<li>They depend on external datums (reference points outside the system)</li>
<li>They cannot be expressed as functions of thermodynamic properties</li>
</ul>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px dashed #a371f7;">
<h3 style="color: #a371f7; margin-top:0;">üß† Memory Trick: The "2 Property Rule"</h3>
<p style="color: #c9d1d9;">
<strong>"Two Friends Fix Everything"</strong>
</p>
<p style="color: #c9d1d9;">
For simple compressible systems, just give me <strong>ANY TWO</strong> independent properties (like P & T, or T & v), and I can tell you <strong>EVERYTHING</strong> about that state!
</p>
<p style="color: #8b949e; font-style: italic;">
It's like unlocking your phone - you need exactly 2 pieces of info (username + password), not more, not less! üîê
</p>
</div>

<h2 style="color: #58a6ff; margin:0;">üåê 4.2 The p-v-T Relation (The 3D Surface)</h2>
</div>

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #58a6ff;">üí≠ The Big Idea</h3>
<p style="color: #c9d1d9;">
We just learned that for simple compressible systems, we need <strong>2 independent properties</strong> to fix the state.
</p>
<p style="color: #c9d1d9; margin-top: 10px;">
From experiments, we found that <strong>specific volume (v)</strong> and <strong>temperature (T)</strong> can be those 2 independent properties.
</p>
<p style="color: #3fb950; font-weight: bold; margin-top: 10px;">
This means: <span style="background-color: #161b22; padding: 5px 10px; border-radius: 5px;">p = p(v, T)</span>
</p>
<p style="color: #8b949e; font-style: italic;">
Pressure is a function of v and T!
</p>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #d29922;">üé® Constructing the 3D p-v-T Surface</h3>
<p style="color: #c9d1d9;">
Let's build this surface step by step:
</p>
<ol style="color: #c9d1d9; line-height: 1.8;">
<li>Draw <strong>v</strong> and <strong>T</strong> as two horizontal axes (x-y plane)</li>
<li>Draw <strong>p</strong> as vertical axis (z-axis)</li>
<li>For each point (v, T) in the horizontal plane, there's a unique pressure p</li>
<li>Plot that point at height p in 3D space</li>
<li>Connect all these 3D points ‚Üí You get a <strong>3D SURFACE</strong>! üéâ</li>
</ol>
</div>




   Each 2D point (v,T) --> One 3D point (v,T,p)
   All 3D points together --> p-v-T SURFACE!
</pre>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #58a6ff;">
<strong style="color: #58a6ff;">üìö CONCEPT: p-v-T Surface</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
A <strong>3D surface</strong> where coordinates of any point represent the values of <strong>pressure, specific volume, and temperature</strong> when the substance is at <strong>equilibrium</strong>.
</p>
</div>

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #3fb950;">üó∫Ô∏è Regions on the p-v-T Surface</h3>

| Region Type | Phases Present | Properties |
|-------------|----------------|------------|
| **Single-Phase** | Solid OR Liquid OR Vapor | Any 2 of (p, v, T) are independent |
| **Two-Phase** | Liquid-Vapor, Solid-Liquid, Solid-Vapor | p & T are NOT independent! |
| **Triple Line** | Solid + Liquid + Vapor | All 3 phases coexist! |

</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #d29922;">
<strong style="color: #d29922;">‚ö†Ô∏è SUPER IMPORTANT!</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
In <strong>Two-Phase Regions</strong>:
</p>
<ul style="color: #c9d1d9; line-height: 1.8;">
<li>Pressure and Temperature are <strong>NOT independent</strong></li>
<li>Change one ‚Üí The other MUST change</li>
<li>Cannot fix state using just p & T</li>
<li>But can fix state using: <strong>v + p</strong> OR <strong>v + T</strong></li>
</ul>
</div>

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #3fb950;">üí° Real-Life Analogy</h3>
<p style="color: #c9d1d9;">
Think of the two-phase region like a <strong>strict relationship</strong>:
</p>
<p style="color: #c9d1d9;">
üîó Temperature and Pressure are in a <strong>committed relationship</strong>. If temperature changes, pressure HAS TO follow. They can't be independent! üòÑ
</p>
<p style="color: #c9d1d9; margin-top: 10px;">
But in single-phase regions, they're <strong>single and independent</strong> - do whatever they want! üòé
</p>
</div>

<strong style="color: #d29922;">üå°Ô∏è CRITICAL POINT - The Boss Point!</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
At the critical point, we have three special values:
</p>
<ul style="color: #c9d1d9; line-height: 1.8;">

‚ö° Above T<sub>c</sub>: Substance exists as vapor and CANNOT be liquified just by increasing pressure!
</p>
<p style="color: #8b949e; font-style: italic;">
To liquify, you MUST bring temperature below T<sub>c</sub> first!
</p>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #58a6ff;">

<li><strong>Pressure at Triple Point</strong> = 0.6113 kPa (measured)</li>
</ul>
<p style="color: #8b949e; font-style: italic; margin-top: 10px;">
At this exact condition, ice, water, and steam all happily coexist together! ‚ùÑÔ∏èüíßüí®
</p>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #58a6ff;">
<strong style="color: #58a6ff;">üìö 2. p-v Diagram</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
Projection of p-v-T surface onto the <strong>pressure-specific volume plane</strong>
</p>
</div>


 ```json                                  
   Key Features:
   - Horizontal lines in vapor dome (isotherms, T = const)
   - Isotherms below Tc: pass through two-phase region
   - Critical isotherm: horizontal at critical point (slope = 0)
   - Isotherms above Tc: continuous decrease, no phase change
```

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #d29922;">üî• Key Observations - p-v Diagram</h3>

| Temperature Range | Behavior as v increases at constant T |
|-------------------|----------------------------------------|
| **T < T<sub>c</sub>** | Pressure constant in two-phase region, decreases in single-phase |
| **T = T<sub>c</sub>** | Critical isotherm, slope = 0 at critical point |
| **T > T<sub>c</sub>** | Pressure continuously decreases, NO two-phase region |

</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #58a6ff;">
<strong style="color: #58a6ff;">üìö 3. T-v Diagram</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
Projection of p-v-T surface onto the <strong>temperature-specific volume plane</strong>
</p>
</div>


```json
   Key Features:
   - Horizontal lines (isobars) in vapor dome (p = const)
   - Isobars below pc: pass through two-phase region
   - Isobars above pc: continuous increase, no phase change
   - P1 < P2 < Pc < P3
```

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #d29922;">üî• Key Observations - T-v Diagram</h3>

| Pressure Range | Behavior as v increases at constant p |
|----------------|----------------------------------------|
| **p < p<sub>c</sub>** | Temperature constant in two-phase region, increases in single-phase |
| **p = p<sub>c</sub>** | Critical isobar |
| **p > p<sub>c</sub>** | Temperature continuously increases, NO two-phase region |

</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px dashed #a371f7;">
<h3 style="color: #a371f7; margin-top:0;">üß† Memory Trick: The Three Diagrams</h3>
<p style="color: #c9d1d9;">
<strong>"P-T, P-V, T-V = The Three Musketeers"</strong>
</p>
<ul style="color: #c9d1d9; line-height: 1.8;">
<li><strong>p-T</strong>: Phase diagram (regions become lines) - Shows WHEN phase changes happen</li>
<li><strong>p-v</strong>: Shows HOW MUCH volume changes during phase change</li>
<li><strong>T-v</strong>: Most commonly used in practice (easiest to work with)</li>
</ul>
<p style="color: #8b949e; font-style: italic; margin-top: 10px;">
Just remember: In two-phase regions, lines are HORIZONTAL (constant p in T-v, constant T in p-v)!
</p>
</div>

<h2 style="color: #d29922; margin:0;">üîÑ 4.4 Phase Change Processes (The Journey!)</h2>
</div>

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #58a6ff;">üé¨ The Experiment Setup</h3>
<p style="color: #c9d1d9;">
Imagine: 1 kg of liquid water at 20¬∞C in a piston-cylinder assembly. We'll heat it slowly at <strong>constant pressure</strong> (1.014 bar) and watch the magic happen! üî•
</p>
</div>



```json    
   Journey: 1(State 1: Initial):(Liquid Region) --> f(Vapor Dome): (Mixture)--> g(Sat. Vapor) --> s(Superheated Vapor)
```

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #3fb950;">üó∫Ô∏è The Complete Journey (Step by Step)</h3>

<table style="width:100%; border-collapse: collapse; margin-top: 10px;">
<tr style="background-color: #0d1117;">
<th style="padding: 10px; border: 1px solid #30363d; color: #58a6ff;">State</th>
<th style="padding: 10px; border: 1px solid #30363d; color: #58a6ff;">Name</th>
<th style="padding: 10px; border: 1px solid #30363d; color: #58a6ff;">What's Happening?</th>
<th style="padding: 10px; border: 1px solid #30363d; color: #58a6ff;">T & v Behavior</th>
</tr>
<tr>
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">1</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Initial Liquid</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Just liquid water at 20¬∞C</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">T = 20¬∞C, v = small</td>
</tr>
<tr style="background-color: #0d1117;">
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">1 ‚Üí f</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Heating Liquid</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Temperature rising, still all liquid</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">T ‚Üë‚Üë, v ‚Üë (slightly)</td>
</tr>
<tr>
<td style="padding: 10px; border: 1px solid #30363d; color: #d29922; font-weight: bold;">f</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Saturated Liquid</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">About to start boiling! 100¬∞C</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">T = 100¬∞C, x = 0</td>
</tr>
<tr style="background-color: #0d1117;">
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">f ‚Üí g</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Two-Phase Mixture</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Boiling! Liquid + Vapor together</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">T constant (100¬∞C), v ‚Üë‚Üë‚Üë</td>
</tr>
<tr>
<td style="padding: 10px; border: 1px solid #30363d; color: #d29922; font-weight: bold;">g</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Saturated Vapor</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Last drop just vaporized! 100¬∞C</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">T = 100¬∞C, x = 1</td>
</tr>
<tr style="background-color: #0d1117;">
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">g ‚Üí s</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Heating Vapor</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Heating steam further</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">T ‚Üë‚Üë, v ‚Üë</td>
</tr>
<tr>
<td style="padding: 10px; border: 1px solid #30363d; color: #d29922; font-weight: bold;">s</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Superheated Vapor</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Hot steam above 100¬∞C!</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">T > 100¬∞C</td>
</tr>
</table>

</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #58a6ff;">
<strong style="color: #58a6ff;">üìö CONCEPT: Subcooled Liquid / Compressed Liquid</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
The region between states 1 and f (before saturation) has two names:
</p>
<ol style="color: #c9d1d9; line-height: 1.8;">
<li><strong>Subcooled Liquid</strong> - Temperature is LESS than saturation temperature at that pressure</li>
<li><strong>Compressed Liquid</strong> - Pressure is HIGHER than saturation pressure at that temperature</li>
</ol>
<p style="color: #8b949e; font-style: italic; margin-top: 10px;">
Both names refer to the SAME states! Use them interchangeably.
</p>
</div>

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #3fb950;">üí° Real-Life Example</h3>
<p style="color: #c9d1d9;">
Water at 80¬∞C and 1.014 bar:
</p>
<ul style="color: #c9d1d9;">
<li><strong>Subcooled</strong> because 80¬∞C < 100¬∞C (saturation temp at 1.014 bar)</li>
<li>It's HOT but not boiling yet! ‚òï</li>
</ul>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #d29922;">
<strong style="color: #d29922;">‚ö° SUPER IMPORTANT: Quality (x)</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
In the two-phase mixture (between f and g), how do we distinguish different states?
</p>
<p style="color: #3fb950; font-weight: bold; margin-top: 10px; font-size: 18px;">
Answer: QUALITY (x)
</p>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #58a6ff;">
<strong style="color: #58a6ff;">üìö CONCEPT: Quality (x)</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
For a two-phase liquid-vapor mixture:
</p>
<p style="color: #3fb950; font-weight: bold; margin-top: 15px; font-size: 18px; text-align: center;">
x = m<sub>vapor</sub> / (m<sub>liquid</sub> + m<sub>vapor</sub>)
</p>
<p style="color: #c9d1d9; margin-top: 15px;">
In simple words: <strong>Quality = Fraction of mass that is vapor</strong>
</p>
</div>

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #3fb950;">üéØ Quality Values at Different States</h3>

| State | Quality (x) | Meaning |
|-------|-------------|----------|
| **Saturated Liquid (f)** | x = 0 | 0% vapor, 100% liquid |
| **Mixture** | 0 < x < 1 | Partly liquid, partly vapor |
| **Saturated Vapor (g)** | x = 1 | 100% vapor, 0% liquid |
| **Subcooled Liquid** | Undefined | No vapor at all! |
| **Superheated Vapor** | Undefined | No liquid at all! |

<p style="color: #8b949e; font-style: italic; margin-top: 10px;">
Quality is ONLY defined in the two-phase region!
</p>
</div>

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #3fb950;">üí° Real-Life Example</h3>
<p style="color: #c9d1d9;">
A mixture with quality x = 0.3 means:
</p>
<ul style="color: #c9d1d9;">
<li>30% of mass is vapor</li>
<li>70% of mass is liquid</li>
</ul>
<p style="color: #8b949e; font-style: italic; margin-top: 10px;">
Think of it like a WhatsApp group: x = 0.3 means 30% people are active (vapor), 70% are silent (liquid)! üòÑ
</p>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #3fb950;">
<strong style="color: #3fb950;">‚úÖ CONCEPT: Superheated Vapor</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
After state g, if we keep heating at constant pressure:
</p>
<ul style="color: #c9d1d9; line-height: 1.8;">
<li>Temperature starts increasing again (above T<sub>sat</sub>)</li>
<li>Specific volume keeps increasing</li>
<li>It's called <strong>Superheated Vapor</strong></li>
</ul>
<p style="color: #c9d1d9; margin-top: 10px;">
<strong>Why "Superheated"?</strong><br>
Because temperature is GREATER than saturation temperature at that pressure!
</p>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #d29922;">
<strong style="color: #d29922;">‚ö†Ô∏è What Happens at Different Pressures?</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
The same heating experiment at different pressures gives different results!
</p>

<table style="width:100%; border-collapse: collapse; margin-top: 10px;">
<tr style="background-color: #0d1117;">
<th style="padding: 10px; border: 1px solid #30363d; color: #58a6ff;">Pressure</th>
<th style="padding: 10px; border: 1px solid #30363d; color: #58a6ff;">What Happens?</th>
</tr>
<tr>
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">p < p<sub>c</sub></td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Normal phase change: Liquid ‚Üí Mixture ‚Üí Vapor<br>Higher p ‚Üí Higher T<sub>sat</sub>, smaller v change</td>
</tr>
<tr style="background-color: #0d1117;">
<td style="padding: 10px; border: 1px solid #30363d; color: #d29922; font-weight: bold;">p = p<sub>c</sub></td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">At critical point - transition happens but no clear distinction</td>
</tr>
<tr>
<td style="padding: 10px; border: 1px solid #30363d; color: #f85149; font-weight: bold;">p > p<sub>c</sub></td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">NO phase change! Continuous transition from liquid-like to vapor-like<br>"Supercritical Fluid"</td>
</tr>
</table>

</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #a371f7;">
<strong style="color: #a371f7;">üåü SPECIAL: Supercritical Fluids</strong>
<p style="color: #c9d1d9; margin-top: 10px;">
When p > p<sub>c</sub> OR T > T<sub>c</sub>, the substance is called a <strong>Supercritical Fluid</strong>.
</p>
<p style="color: #c9d1d9; margin-top: 10px;">
<strong>Naming Convention:</strong>
</p>
<ul style="color: #c9d1d9;">
<li>If T < T<sub>c</sub> ‚Üí Call it "liquid" (for reference)</li>
<li>If T > T<sub>c</sub> ‚Üí Call it "vapor" (for reference)</li>
</ul>
<p style="color: #c9d1d9; margin-top: 10px;">
<strong>Cool Properties:</strong>
</p>
<ul style="color: #c9d1d9;">
<li>Transport properties like gases (can move through porous media)</li>
<li>Can dissolve solutes like liquids</li>
<li>Used in coffee decaffeination, extraction processes! ‚òï</li>
</ul>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px dashed #a371f7;">
<h3 style="color: #a371f7; margin-top:0;">üß† Memory Trick: Phase Changes</h3>
<p style="color: #c9d1d9;">
<strong>"The Temperature Plateau Story"</strong>
</p>
<p style="color: #c9d1d9; margin-top: 10px;">
When phase is changing (f ‚Üí g), temperature takes a <strong>break</strong> (plateau)! üèîÔ∏è
</p>
<p style="color: #c9d1d9;">
All the energy goes into breaking molecular bonds (changing phase), not increasing temperature!
</p>
<p style="color: #8b949e; font-style: italic; margin-top: 10px;">
It's like studying for exams - sometimes you're just stuck at the same level (plateau) even though you're putting in energy! üòÖ
</p>
</div>

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #58a6ff;">üî¨ Why Focus on Liquid-Vapor Phase Change?</h3>
<p style="color: #c9d1d9;">
While melting and sublimation also exist, liquid-vapor phase change is most important in engineering because:
</p>
<ul style="color: #c9d1d9; line-height: 1.8;">
<li><strong>Superior transport properties</strong> - Fluids can flow and carry energy</li>
<li>Can transport <strong>large amounts of heat</strong> efficiently</li>
<li>Used in <strong>coolant applications</strong> (refrigerators, AC, heat exchangers)</li>
<li>Used in <strong>power plants</strong> (steam turbines)</li>
</ul>
<p style="color: #8b949e; font-style: italic; margin-top: 10px;">
Can't really use solids to cool your CPU, right? üòÑ
</p>
</div>

---
<div style="background: linear-gradient(to right, #161b22, #0d1117); padding: 20px; border-left: 5px solid #a371f7; border-radius: 8px;">
<h1 style="color: #a371f7; margin:0;">üìù EXAM CORNER</h1>
<p style="color: #8b949e; margin-top:5px;">Quick Revision Zone - Everything You Need to Ace This!</p>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #58a6ff;">üìä Master Summary Table</h3>

<table style="width:100%; border-collapse: collapse; margin-top: 10px;">
<tr style="background-color: #0d1117;">
<th style="padding: 10px; border: 1px solid #30363d; color: #58a6ff;">Concept</th>
<th style="padding: 10px; border: 1px solid #30363d; color: #58a6ff;">Key Points</th>
<th style="padding: 10px; border: 1px solid #30363d; color: #58a6ff;">Formula/Value</th>
</tr>
<tr>
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">Phase</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Homogeneous in composition & structure</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Solid, Liquid, or Vapor</td>
</tr>
<tr style="background-color: #0d1117;">
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">Pure Substance</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Uniform chemical composition</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Can exist in multiple phases</td>
</tr>
<tr>
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">State Principle</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">For simple systems</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">2 independent properties</td>
</tr>
<tr style="background-color: #0d1117;">
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">p-v-T Surface</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">3D representation of equilibrium states</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">p = p(v, T)</td>
</tr>
<tr>
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">Critical Point</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Top of vapor dome</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">T<sub>c</sub>, p<sub>c</sub>, v<sub>c</sub></td>
</tr>
<tr style="background-color: #0d1117;">
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">Triple Point (Water)</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">All 3 phases coexist</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">273.16 K, 0.6113 kPa</td>
</tr>
<tr>
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">Quality</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Only in two-phase region</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">x = m<sub>vapor</sub>/m<sub>total</sub></td>
</tr>
<tr style="background-color: #0d1117;">
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">Sat. Liquid</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">About to vaporize</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">x = 0, State 'f'</td>
</tr>
<tr>
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">Sat. Vapor</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">Last drop just vaporized</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">x = 1, State 'g'</td>
</tr>
<tr style="background-color: #0d1117;">
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">Subcooled Liquid</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">T < T<sub>sat</sub> at given p</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">= Compressed liquid</td>
</tr>
<tr>
<td style="padding: 10px; border: 1px solid #30363d; color: #3fb950; font-weight: bold;">Superheated Vapor</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">T > T<sub>sat</sub> at given p</td>
<td style="padding: 10px; border: 1px solid #30363d; color: #c9d1d9;">State 's'</td>
</tr>
</table>

</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px dashed #a371f7;">
<h3 style="color: #a371f7; margin-top:0;">üß† Memory Tricks & Mnemonics</h3>

<div style="background-color: #0d1117; padding: 10px; border-radius: 5px; margin-top: 10px;">
<p style="color: #3fb950; font-weight: bold;">1. State Principle</p>
<p style="color: #c9d1d9;">"<strong>Two Friends Fix Everything</strong>" - 2 properties fix state of simple systems</p>
</div>

<div style="background-color: #0d1117; padding: 10px; border-radius: 5px; margin-top: 10px;">
<p style="color: #3fb950; font-weight: bold;">2. Pure Substance</p>
<p style="color: #c9d1d9;">"<strong>Same DNA Everywhere</strong>" - Same composition in all phases</p>
</div>

<div style="background-color: #0d1117; padding: 10px; border-radius: 5px; margin-top: 10px;">
<p style="color: #3fb950; font-weight: bold;">3. Critical Point</p>
<p style="color: #c9d1d9;">"<strong>Critical = Crisis = Identity Crisis</strong>" - Can't distinguish liquid from vapor</p>
</div>

<div style="background-color: #0d1117; padding: 10px; border-radius: 5px; margin-top: 10px;">
<p style="color: #3fb950; font-weight: bold;">4. Quality Formula</p>
<p style="color: #c9d1d9;">"<strong>Quality = Vapor Quotient</strong>" - x = vapor/(liquid + vapor)</p>
</div>

<div style="background-color: #0d1117; padding: 10px; border-radius: 5px; margin-top: 10px;">
<p style="color: #3fb950; font-weight: bold;">5. Saturation States</p>
<p style="color: #c9d1d9;">"<strong>f = First, g = Gone</strong>" - f: first bubble, g: gone (all vapor)</p>
</div>

<div style="background-color: #0d1117; padding: 10px; border-radius: 5px; margin-top: 10px;">
<p style="color: #3fb950; font-weight: bold;">6. Two-Phase Region</p>
<p style="color: #c9d1d9;">"<strong>Temp Takes a Break</strong>" - Temperature constant during phase change</p>
</div>

<div style="background-color: #0d1117; padding: 10px; border-radius: 5px; margin-top: 10px;">
<p style="color: #3fb950; font-weight: bold;">7. The Three Diagrams</p>
<p style="color: #c9d1d9;">"<strong>P-T, P-V, T-V = Three Musketeers</strong>" - Three 2D views of same 3D surface</p>
</div>

</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #d29922;">‚ö° One-Liners for 1-Minute Revision</h3>

<ol style="color: #c9d1d9; line-height: 2.0;">
<li><strong>Phase</strong> = Homogeneous in composition + structure (all solid, all liquid, or all vapor)</li>
<li><strong>Pure substance</strong> = Same chemical composition everywhere, can have multiple phases</li>
<li><strong>State principle</strong> = 2 independent properties fix state of simple systems</li>
<li><strong>Simple compressible</strong> = Energy altered by work only through volume change (‚à´p dV)</li>
<li><strong>p-v-T surface</strong> = 3D surface showing all equilibrium states, p = p(v, T)</li>
<li><strong>Vapor dome</strong> = Dome-shaped two-phase liquid-vapor region on p-v-T surface</li>
<li><strong>Critical point</strong> = Top of dome, max T for liquid-vapor coexistence (T<sub>c</sub>, p<sub>c</sub>, v<sub>c</sub>)</li>
<li><strong>Triple point</strong> = All 3 phases coexist (Water: 273.16 K, 0.6113 kPa)</li>
<li><strong>Saturation state</strong> = State where phase change begins/ends (f or g)</li>
<li><strong>Quality (x)</strong> = Mass fraction of vapor in two-phase mixture (0 to 1)</li>
<li><strong>Subcooled liquid</strong> = T < T<sub>sat</sub> at given p (before f)</li>
<li><strong>Saturated liquid (f)</strong> = About to vaporize, x = 0</li>
<li><strong>Two-phase mixture</strong> = Between f and g, constant T & p, 0 < x < 1</li>
<li><strong>Saturated vapor (g)</strong> = Last drop just vaporized, x = 1</li>
<li><strong>Superheated vapor</strong> = T > T<sub>sat</sub> at given p (after g)</li>
<li><strong>p-T diagram</strong> = Phase diagram, 2-phase regions become lines</li>
<li><strong>p-v diagram</strong> = Horizontal isotherms in vapor dome for T < T<sub>c</sub></li>
<li><strong>T-v diagram</strong> = Horizontal isobars in vapor dome for p < p<sub>c</sub></li>
<li><strong>T<sub>sat</sub> & p<sub>sat</sub></strong> = One-to-one relationship, each p has unique T</li>
<li><strong>Supercritical fluid</strong> = p > p<sub>c</sub> or T > T<sub>c</sub>, no distinct phase change</li>
</ol>

</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #f85149;">üö® Common Mistakes to AVOID!</h3>

<div style="background-color: #0d1117; padding: 10px; border-radius: 5px; margin-top: 10px; border-left: 3px solid #f85149;">
<p style="color: #f85149; font-weight: bold;">‚ùå Mistake 1:</p>
<p style="color: #c9d1d9;">Thinking p & T are independent in two-phase region</p>
<p style="color: #3fb950;">‚úÖ Reality: In two-phase, p & T are linked! Change one ‚Üí other changes too</p>
</div>

<div style="background-color: #0d1117; padding: 10px; border-radius: 5px; margin-top: 10px; border-left: 3px solid #f85149;">
<p style="color: #f85149; font-weight: bold;">‚ùå Mistake 2:</p>
<p style="color: #c9d1d9;">Using quality (x) for subcooled or superheated states</p>
<p style="color: #3fb950;">‚úÖ Reality: Quality ONLY defined between f and g (two-phase region)</p>
</div>

<div style="background-color: #0d1117; padding: 10px; border-radius: 5px; margin-top: 10px; border-left: 3px solid #f85149;">
<p style="color: #f85149; font-weight: bold;">‚ùå Mistake 3:</p>
<p style="color: #c9d1d9;">Thinking you can liquify by just increasing pressure above T<sub>c</sub></p>
<p style="color: #3fb950;">‚úÖ Reality: Above T<sub>c</sub>, no amount of pressure will liquify! Must cool first</p>
</div>

<div style="background-color: #0d1117; padding: 10px; border-radius: 5px; margin-top: 10px; border-left: 3px solid #f85149;">
<p style="color: #f85149; font-weight: bold;">‚ùå Mistake 4:</p>
<p style="color: #c9d1d9;">Confusing triple line with triple point on 2D diagrams</p>
<p style="color: #3fb950;">‚úÖ Reality: 3D surface has triple LINE, but 2D projection shows triple POINT</p>
</div>

</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">
<h3 style="color: #3fb950;">üíé Pro Tips for Exams</h3>

<ul style="color: #c9d1d9; line-height: 1.8;">
<li>üéØ Always identify the <strong>region</strong> first (liquid, two-phase, or vapor)</li>
<li>üéØ In two-phase region, you MUST know quality to fix the state</li>
<li>üéØ Sketch the T-v or p-v diagram - it makes problems 10x easier!</li>
<li>üéØ Remember: f = saturated liquid, g = saturated vapor (f comes before g alphabetically!)</li>
<li>üéØ Horizontal lines in vapor dome: isotherms on p-v, isobars on T-v</li>
<li>üéØ Water is weird - it expands on freezing (most substances contract)</li>
<li>üéØ Subcooled = Compressed (same thing, different names)</li>
<li>üéØ Above critical point, use "liquid" if T < T<sub>c</sub>, "vapor" if T > T<sub>c</sub> (just for reference)</li>
</ul>

</div>

---
<div style="background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); padding: 20px; border-radius: 10px; text-align: center;">
<h2 style="color: #ffffff; margin: 0;">üéâ Part 1 Complete!</h2>
<p style="color: #ffffff; margin-top: 10px; font-size: 18px;">
Awesome job, bhai! You've mastered the fundamentals! üí™
</p>
<p style="color: #ffffff; margin-top: 10px;">
You now understand:<br>
‚úÖ Phases & Pure Substances<br>
‚úÖ State Principle<br>
‚úÖ p-v-T Surface & Projections<br>
‚úÖ Phase Change Processes<br>
</p>
<p style="color: #ffeb3b; margin-top: 15px; font-size: 20px; font-weight: bold;">
Ready for Part 2? Just say "NEXT"! üöÄ
</p>
</div>
