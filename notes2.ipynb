## Question 1

**Define:**

(a) Closed system  
(b) Control volume  
(c) Give one example of each.

<div style='background: #0d1117; border-left: 4px solid #3fb950; padding: 15px; margin: 20px 0;'>

### Solution

**(a) Closed system:** A fixed quantity of matter (fixed mass) where no mass can cross its boundary, but energy can.

**(b) Control volume:** An identifiable region in space through which mass may flow (open system).

**(c) Example:**
- **Closed system:** Gas trapped in a piston-cylinder device with the piston moving but no mass entering/leaving.
- **Control volume:** A water heater with water entering and leaving.

**Marks:** 1 (def each) + 1 (example) = **3 marks**

## Question 2

**Differentiate between macroscopic and microscopic approaches in thermodynamics.**

<div style='background: #0d1117; border-left: 4px solid #3fb950; padding: 15px; margin: 20px 0;'>

### Solution

| Macroscopic (Continuum) | Microscopic (Statistical) |
|--------------------------|----------------------------|
| Deals with gross/average behavior of matter. | Studies molecular structure and interactions. |
| Uses properties like pressure, temperature. | Uses statistics over molecular motions. |
| Assumes matter is continuous. | Considers discrete molecules with gaps. |
| Example: Classical thermodynamics. | Example: Kinetic theory of gases. |

**Marks:** **2 marks**

## Question 3

**State the Zeroth law of thermodynamics and explain its significance.**

<div style='background: #0d1117; border-left: 4px solid #3fb950; padding: 15px; margin: 20px 0;'>

### Solution

**Statement:** If two bodies are each in thermal equilibrium with a third body, they are in thermal equilibrium with each other.

**Significance:** It establishes the basis for temperature measurement and ensures that temperature is a valid thermodynamic property. It allows the use of thermometers to compare temperatures of different systems.

**Marks:** 1 (statement) + 1 (significance) = **2 marks**

## Question 5

**Convert:**

(a) 100¬∞C to Kelvin  
(b) 300 K to ¬∞C  
(c) 68¬∞F to ¬∞C

<div style='background: #0d1117; border-left: 4px solid #3fb950; padding: 15px; margin: 20px 0;'>

### Solution

**(a)** 
$$T_K = T_C + 273.15 = 100 + 273.15 = 373.15 \, \text{K}$$

**(b)** 
$$T_C = T_K - 273.15 = 300 - 273.15 = 26.85^\circ \text{C}$$

**(c)** 
$$T_C = \frac{5}{9}(T_F - 32) = \frac{5}{9}(68 - 32) = \frac{5}{9} \times 36 = 20^\circ \text{C}$$

## Question 6

**The pressure of a gas in a tank is measured as 250 kPa (gage). If the local atmospheric pressure is 100 kPa, find the absolute pressure.**

<div style='background: #0d1117; border-left: 4px solid #3fb950; padding: 15px; margin: 20px 0;'>

### Solution

**Given:** $P_{\text{gage}} = 250 \, \text{kPa}$, $P_{\text{atm}} = 100 \, \text{kPa}$

$$P_{\text{abs}} = P_{\text{gage}} + P_{\text{atm}} = 250 + 100 = 350 \, \text{kPa}$$

**Marks:** **2 marks**

## Question 7

**What is a quasi-static process? Why is it important in thermodynamics?**

<div style='background: #0d1117; border-left: 4px solid #3fb950; padding: 15px; margin: 20px 0;'>

### Solution

A **quasi-static process** is one that occurs slowly enough that the system remains infinitesimally close to thermodynamic equilibrium at every instant.

**Importance:** It allows the system to be described by well-defined thermodynamic properties at each point, making analysis simpler and enabling representation of processes on property diagrams.

**Marks:** 1 (definition) + 1 (importance) = **2 marks**

## Question 2

**State and explain the State Postulate for a simple compressible system. How does it change for a system undergoing phase change?**

<div style='background: #161b22; border-left: 4px solid #58a6ff; padding: 15px; margin: 20px 0;'>

### Solution

**State Postulate:** The state of a simple compressible system is completely specified by **two independent intensive properties**.

**Explanation:** For a single-phase substance (e.g., gas or liquid), knowing any two properties like pressure and temperature fixes all other properties (density, internal energy, etc.).

**For phase change systems** (e.g., water boiling at 100¬∞C), pressure and temperature become **dependent** (related by saturation curve). In such cases, one needs another independent property (e.g., quality or specific volume) along with either P or T to define the state completely.

**Marks:** 2 (postulate) + 2 (explanation) + 1 (phase change) = **5 marks**

</div>

## Question 4

**Explain the difference between absolute, gage, and vacuum pressures. A mercury manometer reads 300 mm of Hg when connected to a tank. If atmospheric pressure is 760 mm Hg, find the absolute pressure in the tank in kPa. (Density of mercury = 13,600 kg/m¬≥, g = 9.81 m/s¬≤)**

<div style='background: #161b22; border-left: 4px solid #58a6ff; padding: 15px; margin: 20px 0;'>

### Solution

**Definitions:**

- **Absolute pressure:** Measured relative to absolute vacuum.
- **Gage pressure:** Difference between absolute and atmospheric pressure.
- **Vacuum pressure:** Pressure below atmospheric (negative gage).

**Given:** $h = 300 \, \text{mm Hg} = 0.3 \, \text{m}$

$$P_{\text{gage}} = \rho g h = 13600 \times 9.81 \times 0.3 = 40024.8 \, \text{Pa} = 40.02 \, \text{kPa}$$

Atmospheric pressure:

$$P_{\text{atm}} = 760 \, \text{mm Hg} = 0.76 \, \text{m}$$

$$P_{\text{atm}} = 13600 \times 9.81 \times 0.76 = 101.3 \, \text{kPa}$$

$$P_{\text{abs}} = P_{\text{gage}} + P_{\text{atm}} = 40.02 + 101.3 = 141.32 \, \text{kPa}$$

**Marks:** 2 (definitions) + 3 (calculation) = **5 marks**

## Question 5

**What is thermodynamic equilibrium? Discuss its types: thermal, mechanical, phase, and chemical equilibrium.**

<div style='background: #161b22; border-left: 4px solid #58a6ff; padding: 15px; margin: 20px 0;'>

### Solution

**Thermodynamic equilibrium:** A state where no unbalanced potentials exist within the system or at its boundaries; properties do not change with time when isolated.

**Types:**

1. **Thermal equilibrium:** Temperature uniform throughout.
2. **Mechanical equilibrium:** No unbalanced forces; pressure uniform (if no other forces).
3. **Phase equilibrium:** Mass distribution among phases remains constant.
4. **Chemical equilibrium:** Chemical composition does not change with time.

**Marks:** 1 (definition) + 4 (four types with brief explanation) = **5 marks**

## Question 6

**With a neat diagram, explain the working principle of a barometer. Derive the expression for atmospheric pressure in terms of mercury column height.**

<div style='background: #161b22; border-left: 4px solid #58a6ff; padding: 15px; margin: 20px 0;'>

### Solution

**Diagram:** 



**Working:** Atmospheric pressure acts on the mercury surface in the reservoir, supporting the mercury column in the tube. At equilibrium, pressure at the same level inside and outside the tube is equal.

**Derivation:**

At point A (surface of reservoir): $P_A = P_{\text{atm}}$

At point B (same horizontal level inside tube): $P_B = P_{\text{vapor}} + \rho g h$

Since $P_A = P_B$:

$$P_{\text{atm}} = P_{\text{vapor}} + \rho g h$$

For mercury, vapor pressure $P_{\text{vapor}}$ is negligible, so:

$$P_{\text{atm}} \approx \rho g h$$
